```python
from langchain_classic.document_loaders import TextLoader 
from langchain_classic.text_splitter import TokenTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_classic.vectorstores import FAISS

#load
loader = TextLoader("data.txt")
document = loader.load()

#split
splitter = TokenTextSplitter(
    chunk_size = 50,
    chunk_overlap=10
)
chunks = splitter.split_documents(documents=document)

#embed
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2",
    model_kwargs = {"device":"cpu"}
)
vectors = embeddings.embed_documents([i.page_content for i in chunks])
#whats the point of turning embeddings to vectors?

#create vector DB
vectorDB = FAISS.from_documents(chunks,embeddings)
```

```python
from langchain_classic.prompts import PromptTemplate
from langchain_classic.schema import SystemMessage,HumanMessage

query="who is my patient?"
similar_docs = vectorDB.similarity_search_with_score(query=query,k=2)
context = []
for doc in similar_docs:
    context.append(doc[0].page_content)

context = "\n".join(context)

temp = """System: Answer in 200 words or less using the context provided.
##Context:{context}
##Prompt:{prompt}"""
prompt_temp = PromptTemplate.from_template(temp)


prompt = prompt_temp.format(context=context,prompt=query)
```

```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
import os

load_dotenv()
llm = ChatOpenAI(
    api_key=os.getenv("API_KEY"),
    base_url="https://openrouter.ai/api/v1",
    model="openrouter/auto",
    temperature=0.7
)
```

```python
loader = TextLoader("text2.txt")
docs = loader.load()
print(f"Loaded {len(docs)} documents")
splitter = RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=30)
chunks = splitter.split_documents(docs)
print(f"Split into {len(chunks)} chunks")
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}: {chunk.page_content[:80]}...")
```

```python
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
vectorestore = FAISS.from_documents(chunks,embeddings)
retriever = vectorestore.as_retriever(search_kwargs={"k":3})
print("vector store ready")
#vector store is the same thing as a vectore DB
```

```python
prompt = ChatPromptTemplate.from_template("""
You are a helpful assistant. Answer the question using Only the context below.
If the answer is not in the context, say "I don't know - that's not in my knowledge base.

Context:
{context}

Question: {question}""")
print(prompt)
```

```python
def format_docs(docs):
    return "\n".join(doc.page_content for doc in docs)

rag_chain = (
    {"context":retriever | format_docs,"question":RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
print(rag_chain)
```

```python
questions = [
    "Who is Oximas?",
    "What is the currency of Dreamland?",
    "How many regions does Dreamland have?",
    "What is the population of Dreamland?",  # not in the doc — should say I don't know
]

for q in questions:
    print(f"Q: {q}")
    answer = rag_chain.invoke(q)
    print(f"A: {answer}")
    print("-" * 40)
```
